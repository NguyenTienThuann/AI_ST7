```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model

url ="/content/insurance.csv"
data = pd.read_csv(url)
data.head()
data.sample(5)

```

```python
#print (data['age'].value_counts())
sns.countplot(x="age",data=data)
```

```python
print (data['smoker'].value_counts())
sns.countplot(x="smoker",data=data)
```

```python
correlation = data.corr() # visualize giữ liệu thành số 
print(correlation)
sns.heatmap(correlation,cmap="Wistia",annot =True) # màu sắc
```

```python
# Đa biến 
model = linear_model.LinearRegression()   
X = data.loc[:,['age','bmi','children']]
model.fit(X , y = data['charges']) # biến y phụ thuộc vào X
a = model.score(X , y = data['charges']) # độ chính xác của mô hình

print('Hệ số intercept là:', model.intercept_)
print('Hệ số coefficient là:', model.coef_)
print('Độ Chính xác của mô hình là:',a )
```

```python
### ĐƠN BIẾN BẬC CAO
model = linear_model.LinearRegression()  # 1 BIẾN, BẬC 10
X = pd.DataFrame([data['age'],data['age']**2,data['age']**3,data['age']**4,data['age']**5,data['age']**6,data['age']**7,data['age']**8,data['age']**9,data['age']**10] ).T #  
X2 = pd.DataFrame(data['age']) # BẬC 1 TEST

model.fit(X, y = data['charges'])
a = model.score(X , y = data['charges']) # độ chính xác của mô hình


print('Hệ số intercept là:', model.intercept_)
print('Hệ số coefficient là:', model.coef_)
print('Độ Chính xác của mô hình là:',a )
```
